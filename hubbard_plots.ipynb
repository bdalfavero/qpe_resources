```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

## $2 \times 2$

```python
f = h5py.File("data/hubbard_out_2_2.hdf5", "r")
eps2_exact = f["eps2_exact"][()]
sample_checkpoints = f["sample_checkpoints"][:]
eps2_samples = f["eps2_samples"][:]
f.close()
```

```python
fig, ax = plt.subplots()
ax.plot(sample_checkpoints, eps2_samples)
ax.hlines(eps2_exact, np.min(sample_checkpoints), np.max(sample_checkpoints), colors="k")
ax.set_xscale("log")
```
