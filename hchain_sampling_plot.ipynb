```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = h5py.File("data/hchain_output_2.hdf5", "r")
sample_checkpoints = f["sample_checkpoints"][:]
sampled_eps2 = f["sampled_eps2"][:]
f.close()
```

```python
fig, ax = plt.subplots()
ax.plot(sample_checkpoints, sampled_eps2)
ax.set_xscale("log")
```

```python
# Plot the self-convergence.
eps2_err = np.abs(sampled_eps2 - sampled_eps2[-1])
fig2, ax2 = plt.subplots()
ax2.plot(sample_checkpoints, eps2_err)
ax2.set_yscale("log")
ax2.set_xscale("log")
ax2.set_ylabel("Error")
ax2.set_xlabel("samples")
```
